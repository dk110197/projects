#scaling the Data

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_sc = sc.fit_transform(X_train)
```

```python
# scaling the test data
X_test_sc = sc.transform(X_test)
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
lr_classifier = LogisticRegression(random_state = 10, penalty='l2')
lr_classifier.fit(X_train_sc,y_train)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score  #for classification report
```

```python
#Train Accuracy:
accuracy_score(y_train, lr_classifier.predict(X_train_sc))

```

```python
CF = confusion_matrix(y_train,lr_classifier.predict(X_train_sc) ) 
  
print('Matrix:')
print(CF) 
```

```python
#optimization of model
from scipy.stats import uniform
from sklearn.model_selection import RandomizedSearchCV

# Create regularization penalty space
penalty = ['l1', 'l2']

# Create regularization hyperparameter distribution using uniform distribution
C = uniform(loc=0, scale=4)

# Create hyperparameter options
hyperparameters = dict(C=C, penalty=penalty)

# Create randomized search 5-fold cross validation and 100 iterations
clf = RandomizedSearchCV(lr_classifier, hyperparameters, random_state=1, n_iter=100, cv=5, verbose=0, n_jobs=-1)

# Fit randomized search
best_model = clf.fit(X_train_sc, y_train)

# View best hyperparameters
print('Best Penalty:', best_model.best_estimator_.get_params()['penalty'])
print('Best C:', best_model.best_estimator_.get_params()['C'])
```

```python
#accuracy score after optimization of model using Random search
accuracy_score(y_train, best_model.predict(X_train_sc))
```
